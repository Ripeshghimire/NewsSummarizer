```python
news['headlines'] = news['headlines'].apply(str.lower)
news['text'] = news['headlines'].apply(str.lower)
news['headlines'] = news['headlines'].apply(str.lower)
news.head()
```

```python
news['news'] = news['headlines'] + ":" + news['ctext']
news['news'][0]
```

```python
news  = news.drop(columns = ['headlines','ctext'],axis = 1)
```

```python
news.rename({"text":"summary"},inplace = True)
```

```python
from transformers import AutoTokenizer
```

```python
checkpoint = 't5-small'
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
from sklearn.model_selection import train_test_split
train_data,temp_data= train_test_split(news,train_size = 0.8,random_state =10)
```

```python
from datasets import Dataset
dataset = Dataset.from_pandas(news)
train_ds = Dataset.from_pandas(train_data, split="train")
test_ds = Dataset.from_pandas(temp_data, split="test")
train_ds
```

```python
prefix = "summarize: "


def preprocess_function(examples):
    inputs = [prefix + doc for doc in examples["news"]]
    model_inputs = tokenizer(inputs, max_length=1024, truncation=True)

    labels = tokenizer(text_target=examples["text"], max_length=128, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
tokenized_train_data = train_ds.map(preprocess_function,batched = True)
tokenized_test_data = test_ds.map(preprocess_function,batched = True)
```

```python
from transformers import DataCollatorForSeq2Seq
data_collator = DataCollatorForSeq2Seq(tokenizer = tokenizer,model = checkpoint)
```

```python
import evaluate
rouge = evaluate.load("rouge")
```

```python
import numpy as np

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    result = rouge.compute(predictions=decoded_preds, references=decoded_labels, use_stemmer=True)

    prediction_lens = [np.count_nonzero(pred != tokenizer.pad_token_id) for pred in predictions]
    result["gen_len"] = np.mean(prediction_lens)

    return {k: round(v, 4) for k, v in result.items()}
```

```python
from transformers import AutoModelForSeq2SeqLM,Seq2SeqTrainingArguments,Seq2SeqTrainer
model  = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir="news_summarization",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=4,
    predict_with_generate=True,
    fp16=True,
    push_to_hub=True,
)
```

```python
trainer = Seq2SeqTrainer(
    model = model,
    args  = training_args,
    train_dataset = tokenized_train_data,
    eval_dataset = tokenized_test_data,
    tokenizer = tokenizer,
    data_collator = data_collator,
    compute_metrics = compute_metrics,
)
```

```python
summarizer = trainer.train()
```
